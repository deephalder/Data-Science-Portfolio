```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold , cross_val_score , train_test_split ,GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
glass = pd.read_csv("Glass.csv")
```

```python
glass.head()
```

```python
glass['Type'].value_counts() #  getting different types of animals
```

```python
# splitting  zoo dataset into x and y 

x = glass.iloc[:,1:-1]
y = glass.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
#using train test split on x and y
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.20, random_state = 42 )
```

### Fitting KNN model

```python
# getting optimum k value score
```

##### Visualizing crossvalidation score

```python
import matplotlib.pyplot as plt 
%matplotlib inline
# choose k between 1 to 41
k_range = range(1, 41)
k_scores = []
# use iteration to caclulator different k in models, then return the average accuracy based on the cross validation
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, x, y, cv=5)
    k_scores.append(scores.mean())
# plot to see clearly
plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```

```python
k = pd.DataFrame({'k_scores': k_scores})
k.sort_values('k_scores',ascending = False)
```

```python
#using n neighbours as 1 to fit model.
```

```python
model = KNeighborsClassifier(n_neighbors=1)
knn = model.fit(x_train, y_train)
results = cross_val_score(model, x_train, y_train, cv=2)
```

```python
results.mean()
```

### predicting x test data

```python
y_pred =model.predict(x_test)
```

```python
results  = pd.DataFrame({'actual':y_test,'Predicted':y_pred})
```

```python
results
```
