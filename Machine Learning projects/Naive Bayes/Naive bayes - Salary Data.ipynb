```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
train = pd.read_csv("SalaryData_Train.csv")
test = pd.read_csv("SalaryData_Test.csv")
```

```python
test.head()
```

```python
df_train = pd.get_dummies(train)
df_train.head()
```

```python
df_train.describe()
```

```python
X = df_train.iloc[:,0:-2]
y = df_train.iloc[:,-1]
```

```python
x_scaled = StandardScaler().fit_transform(X)
```

### Using Gaussian normal distribution Naive bayes algorithm

```python
from sklearn.naive_bayes import GaussianNB
clf_gb =GaussianNB()
model = clf_gb.fit(x_scaled,y)
```

### predicting data with model

```python
df_test = pd.get_dummies(test)
```

```python
X_test = df_test.iloc[:,0:-2]
y_test = df_test.iloc[:,-1]
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

#### Gausian Naive baise algoritm has an output of 78% accuracy.

## ** Using multinomial Naive Bayes therom**

### using MinMaxScaler()

```python
from sklearn import preprocessing
mm_scaler = preprocessing.MinMaxScaler()
X_train_minmax = mm_scaler.fit_transform(X)
```

```python
pd.DataFrame(X_train_minmax).describe()
```

```python
from sklearn.naive_bayes import MultinomialNB as MB
clf_mb =MB()
model_mb = clf_mb.fit(X_train_minmax,y)
```

### predicting test data with MB model

```python
y_pred_1= model_mb.predict(X_test)
accuracy_score(y_pred_1,y_test)
```

#### with MinMaxScaler we get 26% accuracy, which is not good.

### Using original X data now to predict using Multinomial NB

```python
from sklearn.naive_bayes import MultinomialNB as MB
clf_mb1 =MB()
model_mb1 = clf_mb1.fit(X,y)
```

```python
y_pred1= model_mb1.predict(X_test)
accuracy_score(y_pred1,y_test)
```

#### now we get 77.49% accuracy with Multinomial NB.
